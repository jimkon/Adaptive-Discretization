In this demo we will assume that the samples we get between each update are good representative of the actual PDF so we won't have to care about the whole sampling procedure. So we will use the same batch of samples for every update.

```python
space_dimensions = 2
sample_size = 10000
offsets = [0.1, -0.2]
samples = np.random.normal(0.5, 0.05, size=sample_size*space_dimensions).reshape((sample_size, space_dimensions))
#move to the right offsets
samples = samples + np.array(list(offsets for i in range(sample_size)))

plt.hist2d(list(item[0] for item in samples), list(item[1] for item in samples))
plt.colorbar()
plt.show()
```

```python
tree_size = 1365

tree = Tree(space_dimensions, tree_size)
tree.plot()
```

```python
for sample in samples:
    tree.search_nearest_node(sample)
```

```python
plot_values(tree)
```

```python
ME = [tree.get_mean_error()]
size = [tree.get_current_size()]
tree.update()
tree.plot()
```
